```python
import ast
def clean(x):
    s=""
    for i in ast.literal_eval(x):
        s+=i['name'].replace(" ","") + " "
    return s
```

```python
clean("[{'id': 16, 'name': 'Animation'}, {'id': 35, 'name': 'Comedy'}, {'id': 10751, 'name': 'Family hyh'}]")
```

```python
movies['genres'] = movies['genres'].apply(clean)
```

```python
# for extracting only top three members.
def clean_cast(x):
    s=""
    count=0
    try:
        cast_list = ast.literal_eval(x)
        for i in cast_list:
            if count < 3:
                s += i['name'].replace(" ","") + " "
                count += 1
            else:
                break
    except (ValueError, SyntaxError, TypeError):
        # Handle NaN or malformed data
        return ""
    return s.strip()
```

```python
movies['cast'] = movies['cast'].apply(clean_cast)
```

```python
movies['crew'][0]
```

```python
def clean_crew(x):
    s=""
    for i in ast.literal_eval(x):
        if i['job']=='Director':
            s+=i['name'].replace(" ","") + " "
            break
    return s
```

```python
movies['crew'] = movies['crew'].apply(clean_crew)
```

```python
movies['combined_features']=(
    movies['overview'] + ' ' +
    movies['genres'] +
    movies['keywords'] +
    movies['cast'] + ' ' +
    movies['crew']
)
```

```python
df['combined_features'] = df['combined_features'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    if isinstance(text, str):
        text = text.lower()
        text = text.translate(str.maketrans('', '', string.punctuation))
        words = word_tokenize(text)
        words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
        return ' '.join(words)
    else:
        return ''
```

```python
df['combined_features'] = df['combined_features'].apply(preprocess_text)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=10000,stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim.shape
```

```python
indices = pd.Series(df.index, index=df['original_title']).drop_duplicates()
```

```python
cosine_sim
```

```python
def recommend(movie):
    idx = indices[movie]
    sim_score = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_score, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    for i in movie_indices:
        print(df["original_title"].iloc[i])
    return
```

```python
(recommend("Waiting to Exhale"))
```
